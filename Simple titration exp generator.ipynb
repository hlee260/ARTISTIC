```python
import pandas as pd

# Prompt the user for the base multiplier value
Ligand = "VEGF"
K_d_reported = 2

# Define multipliers and row values
multipliers = [0, 1, 2, 2.5, 5, 10, 25, 50, 100]        # Multipliers for Kd
salt_concentation = [0, 5, 10, 25, 50, 100]             # Vertical values


# Create a DataFrame with blanks except for the top row
table = pd.DataFrame(index=salt_concentation, columns=([K_d_reported * mult for mult in multipliers]))
table.index.name = "[KCl] (mM) vs [Ligand] (nM)"

# Save the table to an Excel file
excel_file_path = f"{Ligand} dART titration2.xlsx"
table.to_excel(excel_file_path)

print(f" {Ligand} dART experimental condition saved to {excel_file_path}")
```
